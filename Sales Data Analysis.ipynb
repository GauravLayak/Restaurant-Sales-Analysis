# Analysing the Data

1. Most Prefered Payment Method?

```python
# find unique values in payment column
data['Payment Method'].unique()
```

```python
# how many uniques values in payment column
data['Payment Method'].nunique()
```

```python
# find counts of each unique values
data['Payment Method'].value_counts()
```

```python
# in percentage form
data['Payment Method'].value_counts(normalize=True)*100
```

```python
# Plot bargraph
data['Payment Method'].value_counts().plot(kind='bar')
```

Insights: Credit Card is most prefered method, followed by Cash

2. Most Selling product? (by Quantity, and by Revenue)

```python
# by quantity

data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
# Plot bargraph

# 1st we assign it
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
# 2nd we convert it into dataframe
most_quantity= most_quantity.reset_index()
```

```python
# to check
most_quantity
```

```python
# 3rd we plot
plt.figure(figsize=(9,4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color= ['red', 'blue', 'green', 'yellow', 'violet'])
plt.title('Most Selling Product - By Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.show()
```

```python
# By Revenue

# create a new column
data['Revenue'] = data['Quantity'] * data['Price']
```

```python
# check
data.head()
```

```python
# groupby
data.groupby('Product')['Revenue'].sum().sort_values(ascending= False)
```

```python
# assign it
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending= False)
```

```python
# to check
most_revenue
```

```python
# convert it into dataframe
most_revenue= most_revenue.reset_index()
```

```python
# check
most_revenue
```

```python
# plot bargraph
plt.figure(figsize=(9,4))
plt.bar(most_revenue['Product'], most_revenue['Revenue'], color= ['red', 'blue', 'green', 'yellow', 'violet'])
plt.title('Most Selling Product - By Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue')
```
